## 1. ÏµúÏ†ÅÏùò SVM ÌååÎùºÎØ∏ÌÑ∞ Ï∞æÍ∏∞

### RBF Ïª§ÎÑê - 2Í∞úÏùò ÌååÎùºÎØ∏ÌÑ∞
1. ÎπÑÏö©(C) : ÎßàÏßÑÏùò ÎÑàÎπÑ Ï°∞Ï†ï
2. Í∞êÎßà : Í≤∞Ï†ï Í≤ΩÍ≥ÑÏÑ†Ïùò Í≥°Î•†ÏùÑ Ï°∞Ï†ï

### ÏÇ¨Ïù¥ÌÇ∑Îü∞ - Í∑∏Î¶¨Îìú ÏÑúÏπò(gridsearch) : Í∞ÑÌé∏ÌïòÍ≤å ÏµúÏ†Å ÎπÑÏö©, Í∞êÎßà Ïïå Ïàò ÏûàÏùå

```python
import pandas as pd

df = pd.read_csv('data/csv/basketball_stat.csv')
df
```

```python
# ÏÇ¨Ïù¥ÌÇ∑Îü∞Ïùò train_test_splitÎ°ú Îç∞Ïù¥ÌÑ∞ ÎÇòÎàî
from sklearn.model_selection import train_test_split
```

```python
# Îç∞Ïù¥ÌÑ∞ÏóêÏÑú 20%Î•º ÌÖåÏä§Ìä∏ Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÑÎ•òÌï®
train, test = train_test_split(df, test_size=0.2)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC 
import numpy as np
```

```python
def svc_param_selection(X, y, nfolds):
    svm_parameters =[
        {'kernel': ['rbf'],
         'gamma' : [0.00001, 0.0001, 0.001, 0.01, 0.1, 1],
         'C' : [0.01,0.1,1,10,100,1000]        
        }]
    # ÏÇ¨Ïù¥ÌÇ∑Îü∞ÏóêÏÑú Ï†úÍ≥µÌïòÎäî GridSearchCVÎ•º ÏÇ¨Ïö©Ìï¥ ÏµúÏ†ÅÏùò ÌååÎùºÎØ∏ÌÑ∞Î•º Íµ¨Ìï®
    clf = GridSearchCV(SVC(), svm_parameters, cv=10)
    clf.fit(X_train, y_train.values.ravel())
    print(clf.best_params_)
    
    return clf
```

```python
# 3Ï†êÏäõÍ≥º Î∏îÎ°úÌÇπ ÌöüÏàòÎ•º ÌïôÏäµ Îç∞Ïù¥ÌÑ∞Î°ú ÏÇ¨Ïö©
X_train = train[['3P', 'BLK']]

#ÎÜçÍµ¨ ÏÑ†Ïàò Ìè¨ÏßÄÏÖòÏùÑ ÏòàÏ∏°Í∞íÏúºÎ°ú ÏÑ†Ï†ï
y_train= train[['Pos']]

# ÏµúÏ†ÅÏùò ÌååÎùºÎØ∏ÌÑ∞Î°ú ÌïôÏäµÎêú Î™®Îç∏ÏùÑ clfÎ°ú Ï†ÄÏû•
clf = svc_param_selection(X_train, y_train.values.ravel(), 10)
```

## üëâÏµúÏ†ÅÏùò ÎπÑÏö© C : 1, Í∞êÎßà : 0.1

## 2. Í≤∞Ï†ï Í≤ΩÍ≥ÑÏÑ† ÏãúÍ∞ÅÌôî

```python
#ÏãúÍ∞ÅÌôîÌï† ÎπÑÏö© ÌõÑÎ≥¥Îì§ÏùÑ Ï†ÄÏû•
C_candidates = []
C_candidates.append(clf.best_params_['C'] * 0.01)
C_candidates.append(clf.best_params_['C'])
C_candidates.append(clf.best_params_['C'] * 100)
```

```python
#ÏãúÍ∞ÅÌôîÌï† Í∞êÎßà ÌõÑÎ≥¥Îì§ÏùÑ Ï†ÄÏû•
gamma_candidates= []
gamma_candidates.append(clf.best_params_['gamma'] * 0.01)
gamma_candidates.append(clf.best_params_['gamma'])
gamma_candidates.append(clf.best_params_['gamma'] * 100)
```

```python
# 3Ï†êÏäõÍ≥º Î∏îÎ°úÌÇπ ÌöüÏàòÎ°ú ÌïôÏäµ
X = train[['3P', 'BLK']]
#ÎÜçÍµ¨ÏÑ†Ïàò Ìè¨ÏßÄÏÖòÏùÑ ÌïôÏäµ Î™®Îç∏Ïùò Î∂ÑÎ•òÍ∞íÏúºÎ°ú ÏÇ¨Ïö©
Y = train['Pos'].tolist()
```

```python
#ÏãúÍ∞ÅÌôîÎ•º ÏúÑÌï¥ ÏÑºÌÑ∞(C)ÏôÄ ÏäàÌåÖÍ∞ÄÎìú(SG)Î•º Ïà´ÏûêÎ°ú ÌëúÌòÑ
position=[]
for gt in Y:
    if gt == 'C':
        position.append(0) # ÏÑºÌÑ∞(C) = 0
    else:
        position.append(1) #ÏäàÌåÖÍ∞ÄÎìú(SG) = 1
        
classifiers = []
```

```python
#ÌååÎùºÎØ∏ÌÑ∞ ÌõÑÎ≥¥Îì§ÏùÑ Ï°∞Ìï©Ìï¥ÏÑú ÌïôÏäµÎêú Î™®Îç∏Îì§ÏùÑ Ï†ÄÏû•
for C in C_candidates:
    for gamma in gamma_candidates:
        clf = SVC(C=C, gamma=gamma)
        clf.fit(X,Y)
        classifiers.append((C, gamma, clf))
```

```python
import matplotlib.pyplot as plt
# Í∞Å Î™®Îç∏ÏùÑ ÏãúÍ∞ÅÌôî
plt.figure(figsize=(18,18))
xx, yy = np.meshgrid(np.linspace(0,4,100), np.linspace(0,4,100))

for (k, (C,gamma, clf)) in enumerate(classifiers):
    Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.subplot(len(C_candidates), len(gamma_candidates), k+1)
    plt.title("gamma=10^%d, C=10^%d" % (np.log10(gamma), np.log10(C)), size='medium')
    
    plt.pcolormesh(xx, yy, -Z, cmap=plt.cm.RdBu)
    plt.scatter(X['3P'], X['BLK'], c=position, cmap=plt.cm.RdBu_r, edgecolors='k')
```

--> Í∞ÄÏö¥Îç∞ Ï∞®Ìä∏Í∞Ä gridsearchÎ•º ÌÜµÌï¥ ÏñªÏùÄ CÏôÄ Í∞êÎßàÎ•º ÏÇ¨Ïö©Ìï¥ ÌïôÏäµÌïú SVM Î™®Îç∏ : Í∑†Îì±ÌïòÍ≤å Îëê Í∑∏Î£πÏù¥ Î∂ÑÎ•òÎê®

## 3. Î™®Îç∏ ÌÖåÏä§Ìä∏

```python
X_test = test[['3P', 'BLK']]
y_test = test[['Pos']]

# ÏµúÏ†ÅÏùò ÌååÎùºÎßàÌÑ∞Î°ú ÌïôÏäµÎêú Î™®Îç∏Î°ú ÌÖåÏä§Ìä∏Î•º ÏßÑÌñâ
y_true, y_pred = y_test, clf.predict(X_test)

print(classification_report(y_true, y_pred))
print()
print("accuracy : "+ str(accuracy_score(y_true, y_pred)))
```

```python
# Ïã§Ï†úÍ∞íÍ≥º ÏòàÏ∏°Îêú Í∞í ÎπÑÍµêÌïòÍ∏∞
comparison = pd.DataFrame({'prediction': y_pred, 'ground_truth': y_true.values.ravel()}) #prediction:ÏòàÏ∏°Í∞í, ground_truth:Ïã§Ï†úÍ∞í
comparison
```
